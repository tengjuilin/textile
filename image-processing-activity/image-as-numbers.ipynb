*Step 3: Reading in the image*

The image doesn't exist within the Jupyter Notebook until you actually "read" it into the notebooks data storage. To do that we use the scikit-image command io.imread

Go ahead and check the imports to see that from skimage (scikit-image) we imported the io module

```python
cell_im = io.imread(cell_im_location)
```

*Step 4: Viewing the Image*

For now don't worry about the [0,:,:] next to the image. But focus on the fact that matplotlib is able to show the image using the function 'imshow'.  Where cmap is the color map we choose to apply and it has been set to gray.

Matplotlib automatically includes x and y axis which designate in this case a numerical location for each pixel (0,0 in the top left and 500,500 in the bottom right). 

```python
plt.imshow(cell_im[0,:,:], cmap='gray')
```
